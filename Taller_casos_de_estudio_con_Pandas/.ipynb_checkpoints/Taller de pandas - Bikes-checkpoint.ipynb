# Taller de Pandas

Este conjunto de datos es una lista de cuántas personas había en 7 carriles bici diferentes en Montreal, cada día durante el año 2018.
Fuente open data de Francia http://donnees.ville.montreal.qc.ca/dataset/velos-comptage

```python
#Librerías necesarias
```

```python
import pandas as pd
```

```python
df_raw = pd.read_csv('bikes2018.csv')
```

```python
# Observamos las tres primeras filas del dataset
df_raw[:3]
```

```python
df_raw = pd.read_csv('bikes2018.csv', sep=',', encoding='latin2', parse_dates=['Date'], dayfirst=True, index_col='Date')
df_raw[:3]
```

```python
df_raw['Berri1']
```

```python
# Observamos distintos indicadores estadísticos de los diferentes carriles bici
df_raw.describe()
```

```python
# En el análisis que nos piden debemos analizar principalmente los carriles de Berri, Boyer, Parc, PierDup y Viger
df_selected = df_raw[['Berri1','Boyer','Boyer 2','Parc','PierDup','Viger']]
```

```python
df_selected
```

```python
# Tratamiento de missing data
# Nos indican desde la administración que cuando se obtienen valores nulos significa que no se había habilitado el carril
# Es decir, se trata de un carril nuevo
df_selected.fillna(value = 0)
```

```python
df_selected
```

## Análisis

#### ¿Cuál es el carril más utilizado por los ciudadanos de Montreal?

```python
df_selected.sum()
```

```python
df_selected.sum().max()
```

```python
df_selected.sum().argmax()
```

```python
# Podemos utilizar .plot() para visualizar el sumatorio.
df_selected.sum().plot(kind='bar')
```

#### ¿Cuál es el mes que más ciclistas?

```python
df_selected.reset_index(inplace = True)
```

```python
df = df_selected.copy()
```

```python
df.head()
```

```python
df['Date']
```

```python
df['Year'] = df['Date'].dt.year 
df['Month'] = df['Date'].dt.month 
```

```python
df.head()
```

```python
df.groupby(['Month']).sum()
```

```python
df['Total'] = df['Berri1'] + df['Boyer'] + df['Boyer 2'] + df['Parc'] + df['PierDup'] #+ df['Viger']
```

```python
df.head()
```

```python
df.groupby(['Month']).sum()
```

```python
df[['Month','Total']].groupby(['Month']).sum()
```

```python
df[['Month','Total']].groupby(['Month']).sum().plot(kind='bar')
```
