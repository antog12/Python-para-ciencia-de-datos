## ¿Cual es el tipo de queja más común?

```python
quejas['Complaint Type'].value_counts()[:10]
```

```python
quejas['Complaint Type'].value_counts()[:10].plot(kind='bar')
```

### Selecciona aquellas quejas de Brooklyn y las que se han realizado por el ruido en la calle o las aceras según el campo Complaint Type. ¿Cuántas quejas con estas caracteristicas hay?

```python
quejas[(quejas['Borough'] == "BROOKLYN") & (quejas['Complaint Type'] == "Noise - Street/Sidewalk")][:5]
```
